# Technoeconomic Plotter
This notebook compares cost over time for Lab CFPP, Lab CBPP, and Lab CFPP (with micropad recycling).

```python
import matplotlib.pyplot as plt
import pandas as pd
```

# Lab Calculations

```python
# Load data
lab_cfpp = pd.read_csv(r'./data/raw/CFPP Lab.csv')
lab_cbpp = pd.read_csv(r'./data/raw/CBPP Lab.csv')
lab_cfpp_recycled = pd.read_csv(r'./data/raw/CFPP Lab (recycle mpads).csv')

# column names [time (hours),protein (mg),cost ($)]
```

```python
# for each dataset, add a column of cost in ($ millions)
lab_cbpp['cost ($ millions)'] = lab_cbpp['cost ($)'] / 1_000_000
lab_cfpp['cost ($ millions)'] = lab_cfpp['cost ($)'] / 1_000_000
lab_cfpp_recycled['cost ($ millions)'] = lab_cfpp_recycled['cost ($)'] / 1_000_000
```

```python
# for each dataset add column of cost efficiency for cost/mg at timepoint
lab_cbpp['cost efficiency ($/mg)'] = lab_cbpp['cost ($)'] / lab_cbpp['protein (mg)']
lab_cfpp['cost efficiency ($/mg)'] = lab_cfpp['cost ($)'] / lab_cfpp['protein (mg)']
lab_cfpp_recycled['cost efficiency ($/mg)'] = lab_cfpp_recycled['cost ($)'] / lab_cfpp_recycled['protein (mg)']

lab_cfpp.head()
```

```python
# only keep rows where time (hours) < 1000
lab_cbpp = lab_cbpp[lab_cbpp['time (hours)'] < 1000]
lab_cfpp = lab_cfpp[lab_cfpp['time (hours)'] < 1000]
lab_cfpp_recycled = lab_cfpp_recycled[lab_cfpp_recycled['time (hours)'] < 1000]
```

```python
# Combined plots in one row
fig, axs = plt.subplots(1, 3, figsize=(18, 5))

# Plot 1: Cost ($ millions) over time
axs[0].plot(lab_cbpp['time (hours)'], lab_cbpp['cost ($ millions)'], label='Lab CBPP', color='green')
axs[0].plot(lab_cfpp_recycled['time (hours)'], lab_cfpp_recycled['cost ($ millions)'], label='Lab CFPP (Recycled)', color='orange')
axs[0].plot(lab_cfpp['time (hours)'], lab_cfpp['cost ($ millions)'], label='Lab CFPP', color='blue')
axs[0].set_xlabel('Time (hours)')
axs[0].set_ylabel('Cost ($ millions)')
axs[0].set_title('Total Cost Over Time')
axs[0].legend()
axs[0].grid(True)

# Plot 2: Protein (mg) over time
axs[1].plot(lab_cbpp['time (hours)'], lab_cbpp['protein (mg)'], label='Lab CBPP', color='green')
# axs[1].plot(lab_cfpp_recycled['time (hours)'], lab_cfpp_recycled['protein (mg)'], label='Lab CFPP (Recycled)', color='blue')
axs[1].plot(lab_cfpp['time (hours)'], lab_cfpp['protein (mg)'], label='Lab CFPP (all types)', color='blue')
axs[1].set_xlabel('Time (hours)')
axs[1].set_ylabel('Protein (mg)')
axs[1].set_title('Protein Over Time')
axs[1].legend()
axs[1].grid(True)

# Plot 3: Cost efficiency ($/mg) over time
axs[2].plot(lab_cbpp['time (hours)'], lab_cbpp['cost efficiency ($/mg)'], label='CBPP Lab', color='green')
axs[2].plot(lab_cfpp['time (hours)'], lab_cfpp['cost efficiency ($/mg)'], label='CFPP Lab', color='blue')
axs[2].plot(lab_cfpp_recycled['time (hours)'], lab_cfpp_recycled['cost efficiency ($/mg)'], label='CFPP Lab Recycled', color='orange')
axs[2].set_xlabel('Time (hours)')
axs[2].set_ylabel('Cost Efficiency ($/mg)')
axs[2].set_title('Cost Efficiency Over Time')
axs[2].legend()
axs[2].grid(True)

plt.tight_layout()

plt.subplots_adjust(top=0.88)
plt.suptitle('Lab Scale Comparisons')
plt.show()
```

# Prototype Special Use Case - Switching Protiens

```python
# Load data
proto_cfpp_switch = pd.read_csv(r'./data/raw/CFPP Prototype (Switching Proteins).csv')
proto_cbpp_switch = pd.read_csv(r'./data/raw/CBPP Prototype (Switching Proteins).csv')

# column names [time (hours),protein (mg),cost ($)]
```
