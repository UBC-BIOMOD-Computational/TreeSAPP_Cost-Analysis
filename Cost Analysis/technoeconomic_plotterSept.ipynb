# Technoeconomic Plotter
This notebook compares cost over time for Lab CFPP, Lab CBPP, and Lab CFPP (with micropad recycling).

```python
import matplotlib.pyplot as plt
import pandas as pd
```

# CFPP

```python
# Load data
lab_cfpp = pd.read_csv(r'./data/raw/CFPP Lab.csv')
lab_cbpp = pd.read_csv(r'./data/raw/CBPP Lab.csv')
lab_cfpp_recycled = pd.read_csv(r'./data/raw/CFPP Lab (recycle mpads).csv')

# column names [time (hours),protein (mg),cost ($)]
```

```python
# for each dataset, add a column of cost in ($ millions)
lab_cbpp['cost ($ millions)'] = lab_cbpp['cost ($)'] / 1_000_000
lab_cfpp['cost ($ millions)'] = lab_cfpp['cost ($)'] / 1_000_000
lab_cfpp_recycled['cost ($ millions)'] = lab_cfpp_recycled['cost ($)'] / 1_000_000
```

```python
lab_cfpp.head()
```

```python
# Plot total cost over time for each
plt.figure()
plt.plot(lab_cbpp['time (hours)'], lab_cbpp['cost ($ millions)'], label='Lab CBPP', color='green')
plt.plot(lab_cfpp_recycled['time (hours)'], lab_cfpp_recycled['cost ($ millions)'], label='Lab CFPP (Recycled)', color='orange')
plt.plot(lab_cfpp['time (hours)'], lab_cfpp['cost ($ millions)'], label='Lab CFPP', color='blue')

# add legend and labels
plt.legend()
plt.xlabel('Time (hours)')
plt.ylabel('Cost ($)')
plt.title('Cost Comparison: Lab CFPP vs Lab CBPP vs Lab CFPP (Recycled)')
plt.grid(True)

plt.show()
```

```python
# Plot total cost over time for each
plt.figure()
plt.plot(lab_cbpp['time (hours)'], lab_cbpp['protein (mg)'], label='Lab CBPP', color='green')
# plt.plot(lab_cfpp_recycled['time (hours)'], lab_cfpp_recycled['protein (mg)'], label='Lab CFPP (Recycled)', color='blue')
plt.plot(lab_cfpp['time (hours)'], lab_cfpp['protein (mg)'], label='Lab CFPP (all types)', color='blue')

# add legend and labels*********
plt.legend()
plt.xlabel('protein (mg)')
plt.ylabel('Cost ($)')
plt.title('Cost Comparison: Lab CFPP vs Lab CBPP vs Lab CFPP (Recycled)')
plt.grid(True)

plt.show()
```

```python
# plot cost efficiency (cost per mg protein) over time for each
lab_cbpp['cost efficiency ($/mg)'] = lab_cbpp['cost ($)'] / lab_cbpp['protein (mg)']
lab_cfpp['cost efficiency ($/mg)'] = lab_cfpp['cost ($)'] / lab_cfpp['protein (mg)']
lab_cfpp_recycled['cost efficiency ($/mg)'] = lab_cfpp_recycled['cost ($)'] / lab_cfpp_recycled['protein (mg)']

plt.figure()
plt.plot(lab_cbpp['time (hours)'], lab_cbpp['cost efficiency ($/mg)'], label='CBPP', color='green')
plt.plot(lab_cfpp['time (hours)'], lab_cfpp['cost efficiency ($/mg)'], label='CFPP', color='blue')
plt.plot(lab_cfpp_recycled['time (hours)'], lab_cfpp_recycled['cost efficiency ($/mg)'], label='CFPP Recycled', color='orange')
plt.xlabel('Time (hours)')
plt.ylabel('Cost Efficiency ($/mg)')
plt.title('Cost Efficiency Over Time')
plt.legend()
plt.show()

```
